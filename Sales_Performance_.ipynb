#Analyzing Retail Sales Performance: Insights from Comprehensive Transaction Data

```python
#Import libraries
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
```

## Data Overview

```python
#loading data
df = pd.read_csv('/content/supermarket_sales - Sheet1.csv')
df.head()
```

```python
df.info()
```

## Data Cleaning

- Converted **Date** and **Time** columns to datetime format: Since the Date column is an object type.

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Time'] = pd.to_datetime(df['Time'], format='%H:%M')
```

```python
df.isnull().sum()  # Check for missing values
df.duplicated().sum()  # Check for duplicates
```

**Calculated additional metrics:**

**Profit Margin:** Derived from the gross income and cogs columns.

```python
df['Profit Margin'] = (df['gross income'] - df['cogs']) / df['gross income'] * 100
```

## Data Analysis

### Sales Performance by Branch

```python
sales_by_branch = df.groupby('Branch')['Total'].sum().reset_index()
sales_by_branch
```

### Sales by Gender:

```python
sales_by_gender = df.groupby('Gender')['Total'].sum().reset_index()
sales_by_gender
```

### Top-selling Products:

```python
sales_by_product_line = df.groupby('Product line')['Total'].sum().reset_index()
sales_by_product_line
```

### Customer Type Segmentation:

```python
sales_by_customer_type = df.groupby('Customer type')['Total'].sum().reset_index()
sales_by_customer_type
```

### Rating Analysis:

```python
rating_distribution = df['Rating'].describe()
rating_distribution
```
